```python
# Célula de diagnóstico: lista submodelos 'encoder' e 'decoder' para cada modelo
import os
import tensorflow as tf

for i in range(10):
    path = f'/home/lucas/ConditionalAutoencoder/Modelos/Modelo_Kyoto-{i}/Modelo-Base/Estrutura/Modelo_Kyoto-{i}.keras'
    print('\n--- Modelo', i, '->', path, '---')
    if not os.path.exists(path):
        print('Arquivo não encontrado:', path)
        continue
    try:
        m = tf.keras.models.load_model(path)
    except Exception as e:
        print('Erro ao carregar modelo:', e)
        continue
    # listar submodels (Functional) presentes
    submodels = [layer.name for layer in m.layers if hasattr(layer, 'layers')]
    print('Submodels encontrados:', submodels)
    for subname in ['encoder', 'decoder']:
        if subname in submodels:
            try:
                sub = m.get_layer(subname)
                print(f"\nSummary do submodel '{subname}':")
                sub.summary()
                tf.keras.utils.plot_model(sub, to_file=f'/home/lucas/ConditionalAutoencoder/{subname}_Modelo_Kyoto-{i}.png', show_shapes=True)
            except Exception as e:
                print('Erro ao obter summary de', subname, e)
        else:
            print(f"Submodel '{subname}' não encontrado.")
```
